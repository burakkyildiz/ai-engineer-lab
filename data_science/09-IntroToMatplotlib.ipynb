```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list,weight_list,"g") 
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
#works with numpy
```

```python
np_age_list = np.array(age_list)
```

```python
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
#different examples
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
plt.plot(numpy_arr1, numpy_arr2)
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2, "y*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1,numpy_arr2,"r*-")
plt.subplot(1,2,2) # 1 row, 2 columns, 2nd graph
plt.plot(numpy_arr2,numpy_arr1,"g--")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(numpy_arr1,numpy_arr2,"b*-")
plt.subplot(2,1,2) # 2 row, 1 columns, 2nd graph
plt.plot(numpy_arr2,numpy_arr1,"y--")
plt.show()
```

```python
#customize the axes
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.5,0.5])
figure_axes.plot(numpy_arr1,numpy_arr2,"g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
```

```python
plt.show()
```

```python
#customize labels
```

```python
new_fig = plt.figure(dpi=100)

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_arr1,numpy_arr1 ** 2,label = "numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3, label= "numpy array **3 ")
new_axes.legend(loc=2)
```

```python
plt.show()
```

```python
new_fig.savefig("new_fig.png", dpi=200)
```
